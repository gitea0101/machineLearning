#### 은닝층의 크기를 정의 : hidden_layer_sizes
 - 은닉층의 수와 뉴런의 개수를 튜플로 전달함
 - 10개의 뉴런을 가진 2개의 은닉층 -> hidden_layer_sizes=(10,10)
 - 기본값은 (100,)

#### 활성화 함수 : activation
 - 'activation' : sigmoid : logistic
 - 기본값은 렐루 함수 

#### 규제 강도 : alpha
 - L1 규제는 효과가 크지않아 L2 규제만 지원 (sklearn 한정)
 - 기본값은 0.0001

#### 배치 크기 : batch_size
#### 학습률 초기값 : learning_rate_init
#### 학습률 크기 : learning_rate
#### 학습 횟수 : max_iter

```python
import mglearn
import graphviz
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
```

```python
x, y = make_moons(n_samples=100, noise=0.25, random_state=0)

X_train, X_test, y_train, y_test = train_test_split(x,y,stratify=y, random_state=42)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10), random_state=0)
mlp.fit(X_train,y_train)

plt.figure(facecolor='w')
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```

```python
mlp = MLPClassifier(hidden_layer_sizes=[100,10], max_iter=1000,random_state=0)
mlp.fit(X_train,y_train)

plt.figure(facecolor='w')
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```

```python
plt.figure(figsize=(20,5))
plt.imshow(mlp.coefs_[0], interpolation="none", cmap="viridis")
plt.xlabel("hidden unit")
plt.ylabel("input")
plt.colorbar()
plt.show()
```

```python
mlp = MLPClassifier(hidden_layer_sizes=[100,10], max_iter= 100, activation="tanh",random_state=0)
mlp.fit(X_train,y_train)

plt.figure(facecolor='w')
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```
