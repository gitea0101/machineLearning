```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df["target"] = iris.target
```

```python
df.head()
```

```python
train_df = df.iloc[:,:4]
train_df.head()
```

```python
tsne = TSNE(n_components=2)
tsne_np = tsne.fit_transform(train_df)
```

```python
tsne_df = pd.DataFrame(tsne_np, columns=["component 0", "component 1"])
```

```python
tsne_df.head()
```

t-sne 2차원 데이터 시각화

```python
tsne_df["target"] = df["target"]

tsne_df_0 = tsne_df[tsne_df["target"]==0]
tsne_df_1 = tsne_df[tsne_df["target"]==1]
tsne_df_2 = tsne_df[tsne_df["target"]==2]
```

```python
plt.scatter(tsne_df_0["component 0"], tsne_df_0["component 1"], 
            color = "pink", label ="setosa")
plt.scatter(tsne_df_1["component 0"], tsne_df_1["component 1"], 
            color = "purple", label ="versicolor")
plt.scatter(tsne_df_2["component 0"], tsne_df_2["component 1"], 
            color = "yellow", label ="virginica")

plt.xlabel("component 0")
plt.ylabel("component 1")
plt.legend()
plt.show()
```

3차원 시각화

```python
tsne = TSNE(n_components=3)
tsne_np = tsne.fit_transform(train_df)
```

```python
tsne_df = pd.DataFrame(tsne_np, columns = ["component 0", "component 1", "component 2"])
```

```python
tsne_df.head()
```

```python
tsne_df["target"]=df["target"]
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 3차원 그래프 세팅
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection="3d")

tsne_df_0 = tsne_df[tsne_df["target"]==0]
tsne_df_1 = tsne_df[tsne_df["target"]==1]
tsne_df_2 = tsne_df[tsne_df["target"]==2]

ax.scatter(tsne_df_0["component 0"], tsne_df_0["component 1"],
           tsne_df_0["component 2"], color = "pink", label="setosa")

ax.scatter(tsne_df_1["component 0"], tsne_df_1["component 1"],
           tsne_df_1["component 2"], color = "purple", label="virsicolor")

ax.scatter(tsne_df_2["component 0"], tsne_df_2["component 1"],
           tsne_df_2["component 2"], color = "yellow", label="virginica")

ax.set_xlabel("component 0")
ax.set_ylabel("component 1")
ax.set_zlabel("component 2")
ax.legend()
plt.show()
```
