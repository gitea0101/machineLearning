```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["image.cmap"] = "gray"
```

```python
digits = load_digits()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10,5),
                         subplot_kw = {"xticks" : (), "yticks" : ()})
fig.patch.set_facecolor('w')
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
# PCA
pca = PCA(n_components=2)
pca.fit(digits.data)
```

```python
digit_pca = pca.transform(digits.data)
```

```python
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
          "#A83683", "#4E655E", "#853541", "#3A3120","#535D8E"]

plt.figure(figsize=(10,10))
plt.xlim(digit_pca[:,0].min(), digit_pca[:,0].max())
plt.ylim(digit_pca[:,1].min(), digit_pca[:,0].max())

for i in range(len(digits.data)):
    
    plt.text(digit_pca[i,0], digit_pca[i,1], str(digits.target[i]),
             color = colors[digits.target[i]],
             fontdict= {"weight" : "bold", "size" : 9})
    
plt.xlabel("첫번째 주성분")
plt.ylabel("두번째 주성분")
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
digits_tsne = tsne.fit_transform(digits.data)
```

```python
plt.figure(figsize=(10,10))
plt.xlim(digits_tsne[:,0].min(), digits_tsne[:,1].max()+1)
plt.ylim(digits_tsne[:,1].min(), digits_tsne[:,1].max()+1)

for i in range(len(digits.data)):
    plt.text(digits_tsne[i,0], digits_tsne[i,1], str(digits.target[i]),
             color = colors[digits.target[i]],
             fontdict={"weight" : "bold", "size" : 9})
    
plt.xlabel("t-SNE 특성 0")
plt.ylabel("t-SNE 특성 1")
plt.show()
```
