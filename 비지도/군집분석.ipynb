```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
customer_data = pd.read_csv("shopping-data.csv")
customer_data.head()
```

```python
customer_data.shape
```

```python
data = customer_data.iloc[:,3:6]
```

```python
plt.figure(figsize=(25,7))
plt.title("Customer dendrogram")
dend = dendrogram(linkage(data, method="ward"))
```

```python
fig = plt. figure(figsize=(20,5))

ax1 = fig.add_subplot(1,3,1)
d1 = dendrogram(linkage(data, method="ward"), orientation="left")
ax1 = fig.add_subplot(1,3,2)
d1 = dendrogram(linkage(data, method="ward"), orientation="right")
ax1 = fig.add_subplot(1,3,3)
d1 = dendrogram(linkage(data, method="ward"), orientation="bottom")
```

```python
from sklearn.cluster import AgglomerativeClustering # 계층적 군집분석
from scipy.cluster.hierarchy import dendrogram, linkage # 시각화 특화
```

 - sklearn 은 덴드로그램을 통해서처럼 시각화 할 수 없음
 - 군집이 어떤 과정으로 생성되었는지 시각화하고 싶으면 덴드로그램 사용

```python
cluster = AgglomerativeClustering(n_clusters= 5, linkage="ward")
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=cluster.labels_, cmap="rainbow")
```
