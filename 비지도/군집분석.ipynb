```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
customer_data = pd.read_csv("shopping-data.csv")
customer_data.head()
```

```python
customer_data.shape
```

```python
data = customer_data.iloc[:,3:6]
```

```python
plt.figure(figsize=(25,7))
plt.title("Customer dendrogram")
dend = dendrogram(linkage(data, method="ward"))
```

```python
fig = plt. figure(figsize=(20,5))

ax1 = fig.add_subplot(1,3,1)
d1 = dendrogram(linkage(data, method="ward"), orientation="left")
ax1 = fig.add_subplot(1,3,2)
d1 = dendrogram(linkage(data, method="ward"), orientation="right")
ax1 = fig.add_subplot(1,3,3)
d1 = dendrogram(linkage(data, method="ward"), orientation="bottom")
```
