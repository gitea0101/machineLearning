1. 결측치 처리
 - 제거
 - 채우기

```python
import pandas as pd
import numpy as np
```

```python
d = { "score1" : [100, 90, np.nan, 95],
      "score2" : [30, np.nan, 45, 56],
      "score3" : [52, 40, 80, 98],
      "score4" : [np.nan, np.nan, np.nan, 65]}
```

```python
df = pd.DataFrame(d)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.score1.isnull()]
```

```python
# 행 삭제
df.dropna()
```

```python
df.dropna(axis=0, how="any") # 결측치가 있는 행 전부 
```

```python
# 열 삭제
df.dropna(axis=1)
```

```python
d2 = { "score1" : [100, np.nan, np.nan, 95],
      "score2" : [np.nan, np.nan, np.nan, np.nan],
      "score3" : [52, np.nan, 80, 98],
      "score4" : [np.nan, np.nan, np.nan, 65]}
```

```python
df2 = pd.DataFrame(d2)
df2
```

```python
#행의 전체값이 결측치인 행을 삭제
df2.dropna(how="all")
```

```python
# 임계치 설정해서 제거
df2.dropna(thresh=2)
```

```python
# 특정 열 안에서만 삭제
df.dropna(subset=["score2","score4"])
```

```python
# 결측치 채우기
df.fillna(0)
```

```python
# 이전값으로 채우기
# 앞의 행이 nan 값이면 채워지지않고 그대로 nan값을 가짐
df.fillna(method="pad")
```

```python
# 결측치 뒤의 값으로 채우기
df.fillna(method="bfill")
```

```python
# 결측치를 각 열의 평균값으로 채우기
df.fillna(df.mean())
```

```python
# 각 열의 중간값, 최소값, 최대값
df.fillna(df.median())
```

```python
df.fillna(df.min())
```

```python
df.fillna(df.max())
```

```python
# replace() 로 채우기
df.replace(to_replace=np.nan, value=0)

# replace 함수는 nan값 대신에 특정한 문자를 다른 문자로 바꾸고 싶을때 유용함
```

```python
# interploate() 함수로 채우기, 앞, 뒤 행의 평균값으로 채움
df.interpolate(method="linear", limit_direction="forward")
```

```python
df.interpolate(method="pad", limit=2)
```

### 범주형 데이터를 원핫인코딩으로 변경
 - 모든 데이터를 0과 1로 변환
 - 기계는 모든 데이터를 이진법으로 처리 하기 때문에 레이블인코딩보다는 원핫인코딩 사용

```python
df = pd.read_csv("mushrooms.csv")
df.head()
```

```python
one = pd.get_dummies(df)
one
```

 - get dummies 의 문제점
 
 train 데이터에만 있고 test 데이터에는 없는 카테고리를 원핫인코딩된 컬럼으로 바꿔주지 못함

#### sklearn 의 OneHotEncoder 사용

```python
x = df.iloc[:,1:]
y = df['class']
```

```python
from sklearn.preprocessing import OneHotEncoder

one = OneHotEncoder(sparse=False)
one.fit(x[["cap-shape"]])
```

```python
train_cap = one.transform(x[["cap-shape"]])
train_cap
```

```python
one.categories_
```

```python
o = pd.DataFrame(train_cap, columns=["cap-shape_"+col for col in one.categories_[0]])

o.head()
```

```python
one_x = pd.concat([x.drop(columns="cap-shape"), o], axis=1)
one_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(one_x,y, random_state=0)
```

 ### 범주특성을 레이블 인코딩으로 변환
  - 숫자의 크고 작음에 대한 특성이 작용함
  - 회귀와 같이 연속된 실수를 다루는 알고리즘에서 레이블 인코딩을 사용하면 잘못된 결과값이 나올 수도 있음

```python
one_x["cap-surface"].unique()
```

```python
# cap-ssurface의 유니크한 카테고리 데이터 개수 확인
one_x["cap-surface"].value_counts()
```

```python
one_x["cap-surface"] = one_x["cap-surface"].map({"y":0, "s":1, "f":2, "g":3})
```

```python
one_x["cap-surface"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
encoder.fit(one_x["cap-color"])
```

```python
one_x["cap-color"] = encoder.transform(one_x["cap-color"])
```

```python
one_x["cap-color"].value_counts()
```
